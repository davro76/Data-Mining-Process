```python
# Import libraries, read the dataset, and display the first observations of the dataset
import pandas as pd
filename = "https://raw.githubusercontent.com/fenago/datasets/main/churn.csv"
ChurnData = pd.read_csv(filename,sep=",")
ChurnData.head()
```

```python
# Getting column names
ChurnData.columns
```

```python
# Size of the Dataframe 
ChurnData.shape
```

```python
# Treats the NUMERIC Cols
from sklearn.preprocessing import RobustScaler
rob_scaler = RobustScaler()

# Convert each of the cols to the scaled version
ChurnData['accountlengthScaled'] = rob_scaler.fit_transform(ChurnData['accountlength'].values.reshape(-1,1))
ChurnData['numbervmailmessagesScaled'] = rob_scaler.fit_transform(ChurnData['numbervmailmessages'].values.reshape(-1,1))
ChurnData['totaldayminutesScaled'] = rob_scaler.fit_transform(ChurnData['totaldayminutes'].values.reshape(-1,1))
ChurnData['totaldaychargeScaled'] = rob_scaler.fit_transform(ChurnData['totaldaycharge'].values.reshape(-1,1))
ChurnData['totaldaycallsScaled'] = rob_scaler.fit_transform(ChurnData['totaldaycalls'].values.reshape(-1,1))
ChurnData['totaleveminutesScaled'] = rob_scaler.fit_transform(ChurnData['totaleveminutes'].values.reshape(-1,1))
ChurnData['totalevecallsScaled'] = rob_scaler.fit_transform(ChurnData['totalevecalls'].values.reshape(-1,1))
ChurnData['totalevechargeScaled'] = rob_scaler.fit_transform(ChurnData['totalevecharge'].values.reshape(-1,1))
ChurnData['totalnightminutesScaled'] = rob_scaler.fit_transform(ChurnData['totalnightminutes'].values.reshape(-1,1))
ChurnData['totalnightcallsScaled'] = rob_scaler.fit_transform(ChurnData['totalnightcalls'].values.reshape(-1,1))
ChurnData['totalnightchargeScaled'] = rob_scaler.fit_transform(ChurnData['totalnightcharge'].values.reshape(-1,1))
ChurnData['totalintlminutesScaled'] = rob_scaler.fit_transform(ChurnData['totalintlminutes'].values.reshape(-1,1))
ChurnData['totalintlcallsScaled'] = rob_scaler.fit_transform(ChurnData['totalintlcalls'].values.reshape(-1,1))
ChurnData['totalintlchargeScaled'] = rob_scaler.fit_transform(ChurnData['totalintlcharge'].values.reshape(-1,1))
ChurnData['numbercustomerservicecallsScaled'] = rob_scaler.fit_transform(ChurnData['numbercustomerservicecalls'].values.reshape(-1,1))
# drop the original cols
ChurnData.drop(['accountlength','numbervmailmessages', 'totaldayminutes', 'totaldaycalls',
       'totaldaycharge', 'totaleveminutes', 'totalevecalls', 'totalevecharge',
       'totalnightminutes', 'totalnightcalls', 'totalnightcharge',
       'totalintlminutes', 'totalintlcalls', 'totalintlcharge',
       'numbercustomerservicecalls'],axis=1,inplace=True)

#print
ChurnData.head()
```

```python
# TREAT THE CAT cols (encoding)
ChurnDataCat = pd.get_dummies(ChurnData[['internationalplan','voicemailplan']])
```

```python
ChurnDataCat.head()
```

```python
# Getting column names
ChurnData.columns
```

```python
# Seperating the numerical data
ChurnDataNum = ChurnData [['accountlengthScaled',
       'numbervmailmessagesScaled', 'totaldayminutesScaled',
       'totaldaychargeScaled', 'totaldaycallsScaled', 'totaleveminutesScaled',
       'totalevecallsScaled', 'totalevechargeScaled',
       'totalnightminutesScaled', 'totalnightcallsScaled',
       'totalnightchargeScaled', 'totalintlminutesScaled',
       'totalintlcallsScaled', 'totalintlchargeScaled',
       'numbercustomerservicecallsScaled']]
ChurnDataNum.shape
```

```python
# Merging with the original data frame
# Preparing the X variables
X = pd.concat([ChurnDataCat, ChurnDataNum], axis=1)
print(X.shape)
# Preparing the Y variable
Y = ChurnData['churn']
print(Y.shape)
X.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
# Splitting the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=123,stratify=Y)
# Defining the LogisticRegression function
ChurnDataModel = LogisticRegression()
ChurnDataModel.fit(X_train, y_train)
```

```python
ChurnDataModel = LogisticRegression(max_iter=1000)
ChurnDataModel.fit(X_train, y_train)
```

```python
pred = ChurnDataModel.predict(X_test)
print('Accuracy of Logistic regression model prediction on test set: {:.2f}'.format(ChurnDataModel.score(X_test, y_test)))

# Confusion Matrix for the model
from sklearn.metrics import confusion_matrix
confusionMatrix = confusion_matrix(y_test, pred)
print(confusionMatrix)

from sklearn.metrics import classification_report
print(classification_report(y_test, pred))
```

```python
print('Percentage of positive class :',(y_train[y_train=='Yes'].value_counts()/len(y_train) ) * 100)
print('Percentage of negative class :',(y_train[y_train=='No'].value_counts()/len(y_train) ) * 100)
```

# Undersample

```python
from sklearn.model_selection import train_test_split

# Splitting the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=123,stratify=Y)
```

```python
# let us first join the train_x and train_y for ease of operation

trainData = pd.concat([X_train,y_train],axis=1)
trainData.head()
```

```python
# Finding the indexes of the sample data set where the propensity is 'yes'
ind = trainData[trainData['churn']=='Yes'].index
print(len(ind))

# Seperate the minority classes
minData = trainData.loc[ind]
print(minData.shape)

# Finding indexes of majority class
ind1 = trainData[trainData['churn']=='No'].index
print(len(ind1))
# Seperating the majority class
majData = trainData.loc[ind1]
print(majData.shape)
majData.head()
```

```python
# Take a random sample equal to length of the minority class to make the data set balanced

majSample = majData.sample(n=len(ind),random_state = 123)
print(majSample.shape)
majSample.head()
```

```python
# Concatinating both data sets and then shuffling the data set

balData = pd.concat([minData,majSample],axis = 0)
print('balanced data set shape',balData.shape)

# Shuffling the data set

from sklearn.utils import shuffle

balData = shuffle(balData)
balData.head()
```

```python
balData.shape
```

```python
# Making the new X_train and y_train

X_trainNew = balData.iloc[:,0:19]
X_trainNew.head()

y_trainNew = balData['churn']
y_trainNew.head()
```

```python
# Defining the LogisticRegression function
ChurnModel1 = LogisticRegression(max_iter=1000)
ChurnModel1.fit(X_trainNew, y_trainNew)

# Predicting on the test
pred = ChurnModel1.predict(X_test)
print('Accuracy of Logisticr regression model prediction on test set for balanced data set: {:.2f}'.format(ChurnModel1.score(X_test, y_test)))
```

```python
# Confusion Matrix for the model
from sklearn.metrics import confusion_matrix
confusionMatrix = confusion_matrix(y_test, pred)
print(confusionMatrix)
```

```python
# Confusion Matrix for the model
from sklearn.metrics import confusion_matrix
confusionMatrix = confusion_matrix(y_test, pred)
print(confusionMatrix)

from sklearn.metrics import classification_report
print(classification_report(y_test, pred))
```

```python
# Splitting the data into train and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=0, stratify=Y)

print("Before OverSampling count of yes: {}".format(sum(y_train=='Yes')))
print("Before OverSampling count of no: {} \n".format(sum(y_train=='No')))
```

```python
import smote_variants as sv
import numpy as np

# Instantiating the SMOTE class
oversampler= sv.SMOTE()

# Creating new training set

X_train_us, y_train_us = oversampler.sample(np.array(X_train), np.array(y_train))

# Shape after oversampling

print('After OverSampling, the shape of train_X: {}'.format(X_train_us.shape))
print('After OverSampling, the shape of train_y: {} \n'.format(y_train_us.shape))

print("After OverSampling, counts of label 'Yes': {}".format(sum(y_train_us=='Yes')))
print("After OverSampling, counts of label 'no': {}".format(sum(y_train_us=='No')))
```

```python
# Training the model with Logistic regression model

# Defining the LogisticRegression function

ChurnModel2 = LogisticRegression(max_iter=1000)

ChurnModel2.fit(X_train_us, y_train_us)

# Predicting on the test set
pred = ChurnModel2.predict(X_test)

# Printing accuracy 
print('Accuracy of Logistic regression model prediction on test set for Smote balanced data set: {:.2f}'.format(ChurnModel2.score(X_test, y_test)))

# Confusion Matrix for the model

from sklearn.metrics import confusion_matrix
confusionMatrix = confusion_matrix(y_test, pred)
print(confusionMatrix)

# Classification report for the model

from sklearn.metrics import classification_report
print(classification_report(y_test, pred))
```
